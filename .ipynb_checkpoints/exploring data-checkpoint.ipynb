```python
import pandas as pd
import numpy as np
from scipy.integrate import solve_ivp, solve_bvp
from scipy.optimize import minimize
from matplotlib import pyplot as plt, dates
import datetime
```

```python
cnty_df = pd.read_csv('us-counties.csv')
cnty_df.date = pd.to_datetime(cnty_df.date,format='%Y-%m-%d')
```

```python
utah_mask = cnty_df.state == 'Utah'
salt_lake_mask = cnty_df.county == 'Salt Lake'
utah_cnty_mask = cnty_df.county == 'Utah'
utah_cnty_df = cnty_df[utah_cnty_mask]
salt_cnty_df = cnty_df[salt_lake_mask]
utah_st_df = cnty_df[utah_mask]
```

```python
salt_cnty_df.plot(x='date',y='cases')
plt.show()
```

```python
salt_cnty_df['new_cases'] = np.hstack([np.array([0]), np.diff(salt_cnty_df.cases.values)])
salt_cnty_df['roll'] = salt_cnty_df.new_cases.rolling(window=7).mean()
salt_cnty_df.plot(x='date',y='roll')
plt.show()
```

```python
utah_cnty_df['new_cases'] = np.hstack([np.array([0]), np.diff(utah_cnty_df.cases.values)])
utah_cnty_df['roll'] = utah_cnty_df.new_cases.rolling(window=7).mean()
utah_cnty_df.plot(x='date',y='roll')
utah_cnty_df.plot(x='date',y='new_cases')
utah_cnty_df.plot(x='date',y='cases')
plt.show()
```

```python
sumt_cnty_df = cnty_df[(cnty_df.county == 'Summit') & (utah_mask)]
sumt_cnty_df.plot(x='date', y='cases')
plt.show()
```

```python
new_york_df = cnty_df[cnty_df.county == 'New York City']
# salt_lake_mask = cnty_df.county == 'Salt Lake'
new_york_df.plot(x='date', y='cases')
plt.show()
```

```python
β = .25
γ = 1/10

t0 = 0
tf = 150
points = tf
t = np.linspace(t0,tf,points)
N = 8.399e6
I0 = 52

# this is using a SIR-A where A is asymptomatic. not too different from SIR but interesting
def diff(t, Y):
    S = Y[0]
    I = Y[1]
    R = Y[2]
    dS = -β*S*I
    dI = (β*S*I) - γ*I
    dR = γ*I
    return np.array([dS, dI, dR])


Y0 = np.array([(N-I0)/N, I0/N, 0])

sol = solve_ivp(diff, (t0,tf), Y0, t_eval=t).y

# plt.plot((1-sol[0])[:63])
# plt.plot(new_york_df.cases.values[11:]/N)
# plt.show()
plt.plot(t,sol[0],label='Susceptible',color='black')
plt.plot(t,sol[1],label='Infected',color='red')
plt.plot(t,sol[2],label='Recovered',color='blue')
plt.xlim(0,tf)
plt.ylim(0,1)
plt.title('Modeling Covid-19 in NYC')
plt.xlabel('Days since Mar 11')
plt.ylabel('Proportion of Population')
plt.legend()
plt.savefig('SIR2.pdf')
plt.show()
```

```python
def mini(x):
    α, Tg = x
    t0 = 1
    tf = 63
    points = tf
    t = np.linspace(t0,tf,points)
    N = 8.399e6
    I0 = 52

    def diff(t, Y):
        S = Y[0]
        I = Y[1]
        R = Y[2]
        dS = -(α/t)*I
        dI = (α/t - 1/Tg)*I
        dR = (1/Tg)*I
        return np.array([dS, dI, dR])


    Y0 = np.array([(N-I0)/N, I0/N, 0])

    sol = solve_ivp(diff, (t0,tf), Y0, t_eval=t).y
    return np.linalg.norm((1-sol[0]) - (new_york_df.cases.values[11:]/N))

#  1/10 is based on some articles I found that claim the 
# average length of time someone is sick is 10 days
# the model seems to always take the upper bound
x0 = (3.2, 22.8)

s = minimize(mini, x0, bounds=((0,20),(0,60)))
print(s)
```

```python
# set parameters based on minimization
α, Tg = s.x

# time period
t0 = 1
tf = 150
points = tf
t = np.linspace(t0,tf,points)

# set NYC population and number of initial cases
# this number is based on the number of cases when
# NYC went into lockdown
N = 8.399e6
I0 = 52

# ODE
def diff(t, Y):
    S = Y[0]
    I = Y[1]
    R = Y[2]
#     This is the PGED model introduced in paper
    dS = -(α/t)*I
    dI = (α/t - 1/Tg)*I
    dR = (1/Tg)*I
    
    return np.array([dS, dI, dR])

# set initial cond
Y0 = np.array([(N-I0)/N, I0/N, 0])

sol = solve_ivp(diff, (t0,tf), Y0, t_eval=t).y

# plot the real world data
plt.figure(figsize=(10,4))
plt.plot(new_york_df.cases.values[11:]/N,label='Data')
# plot the modeled data
plt.plot(t[:63],(1-sol[0])[:63],label='1-S(t)')
plt.title('Modeling NYC covid cases with SIR')
plt.legend()
plt.xlabel('Days since Mar 11')
plt.ylabel('Proportion of Population')
plt.savefig('PGED-model.pdf',dpi=500)
plt.show()

plt.plot(t,1-sol[0],label='1-S')
plt.plot(t,sol[1],label='I')
plt.plot(t,sol[2],label='R')
plt.ylim(0,0.15)
plt.legend()
plt.title('Modeling NYC covid cases with SIR')
plt.show()
```
